```python
df = pd.merge(left=df_1, right=df_2, how= "left", on="ISO\ncountry numeric code")

df.columns
```

```python
df = df[["Country_x", 'Multidimensional Poverty Index\n(MPI = H*A)', "Multidimensional Poverty Index\n(MPI)", 'Headcount ratio: Population in multidimensional poverty\n(H)',
       'Intensity of deprivation among the poor\n(A) ', 'Population 2021', "World region_y", 'Health ', 'Education',
       'Living Standards', 'Nutrition', 'Child mortality',
       'Years of schooling', 'School attendance', 'Cooking \nfuel',
       'Sanitation', 'Drinking water', 'Electricity', 'Housing', 'Assets']]

per_country_values = df[["Country_x", "Multidimensional Poverty Index\n(MPI = H*A)", "World region_y", 'Nutrition', 'Child mortality',
       'Years of schooling', 'School attendance', 'Cooking \nfuel',
       'Sanitation', 'Drinking water', 'Electricity', 'Housing', 'Assets', "Population 2021"]]

def standardize(df, x):
    df.loc[:, x] = df["Multidimensional Poverty Index\n(MPI = H*A)"] * df[x] / 100

def get_weights(df):
    total_pop = df["Population 2021"].sum()
    df_copy = df.copy()
    df_copy.loc[:, "Weight"] = df["Population 2021"] / total_pop
    return df_copy
    
indicator_list = [x for x in per_country_values.columns if x not in ['Country_x', 'World region_y',
                                                     "Multidimensional Poverty Index\n(MPI = H*A)",
                                                    "Population 2021"]]

for x in indicator_list:
    standardize(per_country_values, x)
```

```python
per_country_values.rename(columns={"Country_x": "Country",
                                  'Multidimensional Poverty Index\n(MPI = H*A)': "MPI",
                                  "World region_y": "Region",
                                  "Cooking \nfuel": "Cooking fuel"}, inplace= True)
per_country_values.head()
```

```python
regions = list(per_country_values["Region"].unique())
regions
region_dfs = {}

for region in regions:
    df = per_country_values[per_country_values["Region"] == region].drop(columns= "Region")
    region_dfs[region] = df
```

```python
region_dfs.keys()

# region_dfs["Arab States"].columns
```

```python
needs_weights = ['Nutrition', 'Child mortality', 'Years of schooling',
       'School attendance', 'Cooking fuel', 'Sanitation', 'Drinking water',
       'Electricity', 'Housing', 'Assets']

for region in region_dfs.keys():
    var_name = region.lower().replace(" ", "_").replace("-", "_") + "_df"
    weighted_df = get_weights(region_dfs[region])
    
    for col in needs_weights:
        col_name = col + "_w"
        weighted_df.loc[:, col_name] = weighted_df[col] * weighted_df["Weight"]


    output_filepath = f"/Users/samiadam/Desktop/ /School/Y3 S2/IT Project/Filtered Data/{var_name}.csv"
    globals()[var_name] = weighted_df
    globals()[var_name].to_csv(output_filepath)
```
